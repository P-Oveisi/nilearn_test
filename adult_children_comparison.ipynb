First, fMRI data of 10 adults and 10 children are taken from an arbitrary dataset (Richardson et al., 2018):

```python
from nilearn import plotting, image, datasets
ten_adult = datasets.fetch_development_fmri(n_subjects = 10, age_group ='adult')
ten_child = datasets.fetch_development_fmri(n_subjects = 10, age_group ='child')
```

Because each subject's data is a 4D image, I will calculate the mean activation for each subject. The means for the first 5 subjects of of the adult group are plotted:

```python
data_means_adult = []
for i in range (10):
    data_means_adult.append(image.mean_img(ten_adult.func[i]))
    if i <5:
        plotting.plot_epi(data_means_adult[i], title = 'mean activation - adult #{}'.format(i+1))
```

The same is done for the children group:

```python
data_means_child = []
for i in range (10):
    data_means_child.append(image.mean_img(ten_child.func[i]))
    if i <5:
        plotting.plot_epi(data_means_child[i], title = 'mean activation - child #{}'.format(i+1))
```

Next, the mean activation for each group will be calculated. The final results are plotted below:

```python
mean_group_adult = image.mean_img(data_means_adult)
mean_group_child = image.mean_img(data_means_child)
plotting.plot_epi(mean_group_adult, title = 'adult group mean activation')
plotting.plot_epi(mean_group_child, title = 'children group mean activation')
```
